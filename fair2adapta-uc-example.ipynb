```python
import rohub
import uuid
from rdflib import URIRef, BNode, Literal
from pathlib import Path
```

### set F2A RO-Crate profile terms

```python
related_product="https://schema.org/isRelatedTo"
has_type="http://www.w3.org/1999/02/22-rdf-syntax-ns#type"
Product="https://schema.org/Product"
variableMeasured="https://schema.org/variableMeasured"
Variable = "https://w3id.org/iadopt/ont/Variable"
name="https://schema.org/name"
url="https://schema.org/url"
depth="https://schema.org/depth"
value="https://schema.org/value"
temporalCoverage="https://schema.org/temporalCoverage"
spatialCoverage="https://schema.org/spatialCoverage"
Place="https://schema.org/Place"
scenarioId="https://w3id.org/ro/terms/cca/scenarioId"
spatialResolutionAsText="http://data.europa.eu/930/spatialResolutionAsText"
license="https://schema.org/license"
identifier="https://schema.org/identifier"
provenance="http://purl.org/dc/terms/provenance"
```

### login to ROHub

```python
rohub_user = Path("~/rohub_users/rohub-user").expanduser().read_text() #open("~/rohub_users/rohub-user").read().rstrip()
rohub_pwd = Path("~/rohub_users/rohub-pwd").expanduser().read_text() #open("~/rohub_users/rohub-pwd").read().rstrip()
rohub.login(username=rohub_user, password=rohub_pwd)
```

### set identifiers of existing RO and annotation

```python
ro_pid="https://w3id.org/ro-id/ea792c69-9037-4d06-84a8-6fded7356e12"
ro_id = ro_pid.rstrip('/').split('/')[-1]
annotation_id="" #"48153973-acdb-48c4-8785-4ff6cab65429"
```

### create or load RO

```python
if not globals().get("ro_pid"):
    ro_title="Arctic Radioisotopes - Sea surface observations"
    ro_research_areas=["Radiobiology"]
    ro_description="This Research Object aggregates the research resources related to sea surface observations"
    #ro_type="Data Research Object"
    #ro = rohub.ros_create(title=ro_title, research_areas=ro_research_areas, description=ro_description, ros_type=ro_type)
    ro = rohub.ros_create(title=ro_title, research_areas=ro_research_areas, description=ro_description)
else:
    ro = rohub.ros_load(identifier=ro_id)
    
```

### create or load annotation

```python
if not globals().get("annotation_id"):
    new_annot=ro.add_annotations()
    annotation_id = new_annot['identifier']
print ("annotation_id: ", annotation_id)
```

### set identifiers for RO resources

```python
# define MVP
mvp_id="https://w3id.org/ro-id/"+ro_id+"/product/1"
# define variable(s)
variable_id="https://w3id.org/ro-id/"+ro_id+"/variable/1"
variable_id2="https://w3id.org/ro-id/"+ro_id+"/variable/2"
# define spatialCoverage entity(ies)
spatialCoverage_id="https://w3id.org/ro-id/"+ro_id+"/spatial/1"
# define value(s)
depth_value_id="https://w3id.org/ro-id/"+ro_id+"/value/"+str(uuid.uuid4())
```

### set values 

```python
depth_value="0-1 m"
temporalCoverage_value="2010/2018"
scenarioId_value = "JRAOC20TRNRPv2"
spatialResolutionAsText_value = "1 degree"
license_url_value="https://creativecommons.org/licenses/by/4.0/"
mvp_id_value="" #"10.2.2/344"
provenanve_url_value="" #"https://example.org/provenance"
variable_name_value="Sea surface salinity (SSS)"
variable_url_value="https://vocab.nerc.ac.uk/standard_name/sea_surface_salinity/"
variable2_name_value="Sea surface temperature (SST)"
variable2_url_value="https://vocab.nerc.ac.uk/standard_name/sea_surface_temperature/"
spatialCoverage_name_value="Earth"
spatialCoverage_url_value="https://www.geonames.org/6295630"
```

### create all the annotations

```python
# connect mvp to RO-Crate
the_subject=ro_pid
the_predicate=related_product
the_object=mvp_id
ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

# set mvp type
the_subject=mvp_id
the_predicate=has_type
the_object=Product
ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

# set mvp associated variable(s)
## variable1
if variable_name_value != "":
    the_subject=mvp_id
    the_predicate=variableMeasured
    the_object=variable_id
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=variable_id
    the_predicate=has_type
    the_object=Variable
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=variable_id
    the_predicate=name
    the_object=variable_name_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

    the_subject=variable_id
    the_predicate=url
    the_object=variable_url_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

## variable2
if variable2_name_value != "":
    the_subject=mvp_id
    the_predicate=variableMeasured
    the_object=variable_id2
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=variable_id2
    the_predicate=has_type
    the_object=Variable
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=variable_id2
    the_predicate=name
    the_object=variable2_name_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

    the_subject=variable_id2
    the_predicate=url
    the_object=variable2_url_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")


# set mpv associated depth
if depth_value != "":
    the_subject=mvp_id
    the_predicate=depth
    the_object=depth_value_id
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=depth_value_id
    the_predicate=value
    the_object=depth_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

# set mvp associated temporal coverage
if temporalCoverage_value != "":
    the_subject=mvp_id
    the_predicate=temporalCoverage
    the_object=temporalCoverage_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

# set mvp associated spatial coverage
if spatialCoverage_name_value != "":
    the_subject=mvp_id
    the_predicate=spatialCoverage
    the_object=spatialCoverage_id
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=spatialCoverage_id
    the_predicate=has_type
    the_object=Place
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

    the_subject=spatialCoverage_id
    the_predicate=name
    the_object=spatialCoverage_name_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

    the_subject=spatialCoverage_id
    the_predicate=url
    the_object=spatialCoverage_url_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

# set mvp associated scenarioId
if scenarioId_value != "":
    the_subject=mvp_id
    the_predicate=scenarioId
    the_object=scenarioId_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

# set mvp associated spatial resolution as text
if spatialResolutionAsText_value != "":
    the_subject=mvp_id
    the_predicate=spatialResolutionAsText
    the_object=spatialResolutionAsText_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

# set mvp associated license
if license_url_value != "":
    the_subject=mvp_id
    the_predicate=license
    the_object=license_url_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id, object_class="URIRef")

# set mvp associated identifier
if mvp_id_value != "":
    the_subject=mvp_id
    the_predicate=identifier
    the_object=mvp_id_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

#
if provenanve_url_value != "":
    the_subject=mvp_id
    the_predicate=provenance
    the_object=provenanve_url_value
    ro.add_triple(the_subject=the_subject, the_predicate=the_predicate, the_object=the_object, annotation_id=annotation_id)

print("task completed")
```
